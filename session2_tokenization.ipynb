🔹 What is Tokenization in NLP?

Tokenization = breaking a sentence or text into smaller pieces called tokens.

👉 Example:
"I love coding" → ["I", "love", "coding"]

These tokens help computers understand text.

🔹 Types of Tokenization (from your screen)
1. Basic Tokenization

Whitespace → split by spaces
"I love NLP" → ["I", "love", "NLP"]

Word Tokenization → splits into words (most common)

Sentence Tokenization → splits into sentences
"Hi. How are you?" → ["Hi.", "How are you?"]

Character Tokenization → splits into letters
"Hi" → ["H", "i"]

Regex Tokenization → uses patterns (advanced splitting)

Penn Treebank → handles punctuation smartly
"don't" → ["do", "n't"]

2. Sequence Generation

N-grams → group of N words
"I love coding"

Bigrams → ["I love", "love coding"]

3. Modern Subword Tokenization

Breaks words into smaller meaningful parts
"unhappiness" → ["un", "happi", "ness"]

👉 Used in models like ChatGPT, BERT, etc.

🔥 Why Tokenization is Important?

Makes text understandable for machines

Helps in:

Sentiment analysis

Chatbots

Translation

Search engines

## Tokenization with Newline (\n)
Newline acts as whitespace, so it splits the text.

```python
from nltk.tokenize import WhitespaceTokenizer

text = "NLTK is a powerful library. It is used for NLP tasks."
tokenizer = WhitespaceTokenizer()

tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
name = "Manan Mishra"
tokens = tokenizer.tokenize(name)

print(tokens)
```

```python
name = "Manan Mishra\nStudent-CSAIML"
tokens = tokenizer.tokenize(name)

print(tokens)
```

```python
name = "Manan\tMishra\tStudent-CSAIML"
tokens = tokenizer.tokenize(name)

print(tokens)
```

## Word Tokenizer

Splits text into words and punctuation using English grammar rules.

```python
from nltk.tokenize import word_tokenize

text = "NLTK is a powerful library. It is used for NLP tasks."
words = word_tokenize(text)

print(words)
```

```python
name = "Manan Mishra"
words = word_tokenize(name)

print(words)
```

```python
text = "Don't hesitate to ask."
words = word_tokenize(text)

print(words)
```

```python
text = "Don't hesitate to buy if price is $100"
words = word_tokenize(text)

print(words)
```

## Sentence Tokenizer

Splits text into sentences using punctuation like (. ! ?)
and understands common abbreviations.

```python
from nltk.tokenize import sent_tokenize

text = "NLTK is a powerful library. It is used for NLP tasks."
sentences = sent_tokenize(text)

print(sentences)
```

```python
name = "Manan Mishra"
sentences = sent_tokenize(name)

print(sentences)
```

```python
name = "Manan Mishra. Student-CSAIML"
sentences = sent_tokenize(name)

print(sentences)
```

```python
text = "Miss. Hello."
sentences = sent_tokenize(text)

print(sentences)
```

```python
text = "Ms. Hello."
sentences = sent_tokenize(text)

print(sentences)
```

```python
text = "Mrs. Hello."
sentences = sent_tokenize(text)

print(sentences)
```

```python
text = "Mr. Hello."
sentences = sent_tokenize(text)

print(sentences)
```

## Regular Expression Tokenizer

Uses user-defined patterns (regex) to split text.
This is more flexible and powerful than other tokenizers.

```python
from nltk.tokenize import regexp_tokenize

text = "Email me at test@example.com"
tokens = regexp_tokenize(text, pattern=r'\w+')

print(tokens)
```

```python
text = "Hey! How's life?"
tokens = regexp_tokenize(text, pattern=r'\w+')

print(tokens)
```

```python
text = "Hey! How's life?"
tokens = regexp_tokenize(text, pattern=r'[A-Za-z]+')

print(tokens)
```

## Treebank Tokenizer

Follows Penn Treebank rules.  
It properly separates:
- Contractions (don't → do + n't)
- Punctuation

```python
from nltk.tokenize import TreebankWordTokenizer

text = "Don't hesitate to ask."
tokenizer = TreebankWordTokenizer()

tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
text = "Don't hesitate to buy if price is $100."
tokenizer = TreebankWordTokenizer()

tokens = tokenizer.tokenize(text)
print(tokens)
```
